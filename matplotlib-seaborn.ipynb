```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.info()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(10,5))

sns.histplot(tips.tip, ax=ax[0], kde=True) # auto bins
sns.histplot(tips.tip, bins=10, ax=ax[1], kde=True)
sns.histplot(tips.tip, bins=100, ax=ax[2], kde=True)
sns.histplot(tips.tip, bins=1000, ax=ax[3], kde=True)

plt.tight_layout()
```
